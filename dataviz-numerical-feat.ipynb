# Data Visualization for Numerical Features

#### Import Packages

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

#### Read data into DataFrame

```python
df = pd.read_csv('heart.csv')
```

#### Visualization of numerical features

```python
df.describe().columns
```

```python
sns.distplot(df.Age)
```

```python
sns.distplot(df.RestingBP)
```

```python
sns.distplot(df.Cholesterol)
```

```python
sns.histplot(df.FastingBS)
```

```python
sns.distplot(df.MaxHR)
```

```python
sns.distplot(df.Oldpeak)
```

```python
sns.lmplot(x='Age',y='MaxHR', data = df, hue = 'Sex')
plt.show()
```

```python
sns.stripplot(x='FastingBS',y='Age', data = df,size =5, jitter = True)
plt.show()
```

```python
sns.swarmplot(x='FastingBS',y='Age',data = df, hue = 'Sex')
plt.show()
```

```python
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# the first plot
sns.boxplot(x='FastingBS',y='Age', data = df)

plt.subplot(1,2,2)

# the second plot
sns.violinplot(x='FastingBS',y='Age', data = df)
plt.tight_layout()
plt.show()
```
